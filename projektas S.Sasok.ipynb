# 1.3. Regionas, kuris greičiausiai vystėsi pagal HDI.
### 1.3.1. Rasime duomenys su regionais ir šalimis ir sujungsime dvi lenteles.

```python
regions = pd.read_html('https://meta.wikimedia.org/wiki/List_of_countries_by_regional_classification')
```

```python
regions
```

```python
print(type(regions))
```

```python
region = pd.DataFrame(np.array(regions).reshape(-1, 3))
```

```python
region = region.rename(columns={0:'Country', 1:'Region', 2:'Global'})
```

```python
hdi1 = pd.read_csv('HDI.csv', index_col=0, na_values=['..'])
```

```python
for_regions = pd.merge(hdi1,region, on=['Country'])
```

```python
for_regions
```

### 1.3.2. Vidutinę rodiklio reikšmę pagal regionus 2019 m.
Taip sužinosime regionų paskirstimą pagal šį rodyklį 2019 m.

```python
regionai_2019 = (for_regions.groupby(['Region'])['2019'].mean()).sort_values(ascending = False)
```

```python
regionai_2019
```

### 1.3.3. Vystymo apibrežimas
Didžiausias vidutinis HDI skirtumas tarp 1990 ir 2019 metais parodys didžiausią vystymą.

```python
vystymas = ((for_regions.groupby(['Region'])['2019'].mean())-(for_regions.groupby(['Region'])['1990'].mean())).sort_values(ascending = False)
```

```python
vystymas
```

##### Stipriausi regionai turi mažiausią vystymo lygį, nes iš pradžių jie turėjo aukštą HDI lygį. Silpniausi regionai turi aukštesnį išsivystymo lygį.

# 2. Analizė pagal GDI komponentus

GDI - (Gender Development Index).
GDI įvertina lyčių skirtumus žmogaus vystymosi pasiekimuose, įvertindama moterų ir vyrų skirtumus trijose 
pagrindinėse žmogaus raidos dimensijose – sveikatos, žinių ir gyvenimo lygio, naudojant tuos pačius komponentų 
rodiklius kaip ir HDI.

```python
gdi = pd.read_csv('GDI.csv',skiprows=[1],index_col=0,na_values=['..'],decimal='.',thousands=',') ##duomenys per 2019
```

```python
gdi
```

#### TOP 10  šalys pagal HDI

```python
top10 = gdi[gdi.index <11]
```

```python
top10
```

## 2.1. Gyvenimo trukmė

##### Duomenis apie gyvenimo trukmę tarp 10 geriausių šalių

```python
top10.plot.bar(x='Country',y=['Lif_Expec_Female','Lif_Excep_Male'], ylim = (78,89),figsize=(12,3),title='Top 10 šalių')
```

## 2.2. Išsilavinimas

#### 2.2.1. Numatoma mokymosi trukmė
##### 10 šalių, kur vyrai turi  daugiausiai mokytis 

```python
gdi.groupby(['HDI Rank','Country'])['Excep_Yrs_Schooling_Male'].max().sort_values(ascending = False)[:10]
```

##### 10 šalių, kur moterys turi  daugiausiai mokytis 

```python
gdi.groupby(['HDI Rank','Country'])['Excep_Yrs_Schooling_Female'].max().sort_values(ascending = False)[:10]
```

##### 10 šalių, kuriose moterų ir vyrų išsilavinimo trukmės skyrtumas yra didžiausias

```python
mok_skyrt = abs(gdi['Excep_Yrs_Schooling_Female']-gdi['Excep_Yrs_Schooling_Male']) 
```

```python
gdi1=gdi.assign(Excep_Yrs_Schooling_Diff = mok_skyrt)
```

```python
gdi1.groupby(['HDI Rank','Country','Excep_Yrs_Schooling_Female','Excep_Yrs_Schooling_Male'])['Excep_Yrs_Schooling_Diff'].max().sort_values(ascending = False)[:10]
```

Šį skirtumą gali lemti kultūra. Nėra bendros tendencijos, kai kur vyrai mokosi ilgiau, o kai kur moterys.

##### Moterų ir vyrų mokymosi metų skirtumas 10 geriausių šalių

```python
gdi1[['Country','Excep_Yrs_Schooling_Female','Excep_Yrs_Schooling_Male','Excep_Yrs_Schooling_Diff']][:10]
```

Beveik visose šalyse moterys turi studijuoti daugiau. Daugiausią skirtumą turi Skandinavijos šalys. Vyrai turi studijuoti daugiau tik: Šveicarijoje ir Vokietijoje.
